```python
import random

alreadyWritten = []
unknow = []
dictionary=["dog","cat","fish","PARROT"]
word = []
lives = 7
position = 0
choice = ""

#the hangman art

hangman = [ "\n ____\n|/   |\n|   \n|    \n|    \n|    \n|\n|_____",
"\n ___\n|/   |\n|   (_)\n|    \n|    \n|    \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|    |\n|    |    \n|    \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|   \|\n|    |\n|    \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|   \|/\n|    |\n|    \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|   \|/\n|    |\n|   / \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|   \|/\n|    |\n|   / \ \n|\n|_____",
"\n ____\n|/   |\n|   (_)\n|   /|\ \n|    |\n|   | |\n|\n|_____"]

#defining a random word

def select():
    global dictionary
    global word
    global unknow
    global choice
    
    choice = random.choice(dictionary)
    if str.isalnum(choice):
        word = list(choice.lower())
        
        #when the game is restarted, the unknow setted to nothing
        
        unknow = []
        
        for i in range(len(word)):
                
            unknow.append("_")
            
    else:
        select()
        
def view():
    global position
    
    print(hangman[position])
    
#a letter is entered, and it is checked that it had not been used previously    

def write():
    global alreadyWritten
    global word
    global unknow
    global position
    global lives
    global choice
    check = False
    
    while check == False:
        
        letter = str(input("Write a letter"))
        
        if len(letter) < 2:
            if letter.lower() in alreadyWritten:
                
                print("You already tried")
            
            else:
                
                check = True
                alreadyWritten.append(letter.lower())
                
                for i in range(len(word)):

                    if letter.lower() == word[i].lower():
                        unknow[i] = letter.lower()
                
                if letter.lower() not in choice.lower():
                    lives = lives -1
                    position= position + 1
        else:
            print("Write a letter not a word")
            
        view()
        print(" ".join(unknow))

#victory and defeat condition        

def countdown():
    global lives
    global unknow
    global word
    
    view()
    print("Guess the word\n" + " ".join(unknow))
    
    while lives > 0:
        
        if unknow != word:
            write()
            
        elif unknow == word:
            print("Well done")
            
            play()

#a difficulty selector           

def diff():
    global word
    global alreadyWritten
    global unknow
    
    print("Select the difficulty\n\n 0.Normal\n 1.Hard\n 2.Back")
            
    difficulty = int(input())

    if difficulty in range(3):

        if difficulty == 0:

            first = word[0]
            last = word[-1]


            for i in range(len(word)):

                if last.lower() == word[i]:

                    unknow[i] = last.lower()

                    alreadyWritten.append(last.lower())
                
                elif first.lower() == word[i]:

                    unknow[i] = first.lower()

                    alreadyWritten.append(first.lower())

            countdown()

        elif difficulty == 1:

            countdown()

        elif difficulty == 2:

            play()
    else:

        print("Select an option please")
        
        diff()
        
#the game incitu    
    
def play():
    global unknow
    global word
    global alreadyWritten 
    
    alreadyWritten = []
    
    select()
    
    print("Select an option\n\n 0.Instructions\n 1.Play\n 2.Quit")
    
    option = int(input())
    
    if option in range(2):
        if option == 0:
            
            print("The object of hangman is to guess the secret word before the stick figure is hung.\nPut a letter, if it is not in the hidden word, a part will be added to the hangman's body\nIf the letter is part of the word it becomes visible\nIf the word is completed, you win ")
            play()
            
        elif option == 1:
            
            diff()
            
        elif option == 2:
            return(0)
```
